# Ground Sample Distance calculation

```python
import math
import matplotlib.pyplot as plt
import numpy as np
from decimal import Decimal
from ipywidgets import widgets
from ipywidgets import interact, interactive, fixed, interact_manual


#Alt = float(input("Enter Alt "))
lbl1=widgets.Label('Enter Alt')
display(lbl1)
text1=widgets.Text()
display(text1)

#Horz = float(input("Enter Horziontal "))
lbl2=widgets.Label('Enter Horizontal')
display(lbl2)
text2=widgets.Text()
display(text2)

#Vert = float(input("Enter Vertical "))
lbl3=widgets.Label('Enter Vertical')
display(lbl3)
text3=widgets.Text()
display(text3)

#p = float(input("Enter Pixel Size "))
lbl4=widgets.Label('Enter Pixel Size')
display(lbl4)
text4=widgets.Text()
display(text4)

#EFL = float(input("Enter Lens Focal Length "))
lbl5=widgets.Label('Enter Lens Focal Length')
display(lbl5)
text5=widgets.Text()
display(text5)

#EFD = float(input("Enter Lens Aperature "))
#lbl6=widgets.Label('Ente Lens Aperature')
#display(lbl6)
#text6=widgets.Text()
#display(text6)

#OffNadirAngle = float(input("Enter Off "))
lbl7=widgets.Label('Enter Off Nadir Angle')
display(lbl7)
text7=widgets.Text()
display(text7)

btn=widgets.Button(description='Calculate')
display(btn)
lbl3=widgets.Label()
display(lbl3)

def calc (b):
    Alt=float(text1.value)
    Horz=float(text2.value)
    Vert=float(text3.value)
    p=float(text4.value)
    EFL=float(text5.value)
    #EFD=float(text6.value)
    ONA=float(text7.value)

    theta = p/EFL
    gsd = theta*Alt
    print ('gsd = %5.2e' % (gsd),'M*2')
    flvw = Horz*Vert*gsd
    print ('Full Field Of View %5.2e' %(flvw))
    
    H= Alt/math.cos(ONA)
    print ('H = %5.2e' %(H))
    
    Angle = 300
    maxx = 20
    minx= 0
    maxy = 1000
    miny =0
    x = np.arange(minx, maxx,1)
    def plot_func(Angle):
        colr='blue'
        plt.plot (x , maxy - abs(900 - Angle)*(x),color = colr)
        plt.plot (x , maxy - (abs(900 - Angle) + 100)*(x), color = 'red')
        plt.axis([ minx, maxx, miny, maxy])
        plt.text(10, 200, "  Theta =" + str(Angle) + "degrees", color='black')
    
    interact(plot_func, Angle = widgets.FloatSlider(value=350, min=10, max=1000, step=10))
    
btn.on_click(calc)
```
